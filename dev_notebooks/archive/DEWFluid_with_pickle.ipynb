# Testing the DEWFluid 'coder' module - passing a user-made system
Ideally the DEW system should be user defined, which means it must be passed to the DEWFluid module before it is initialised as a phase.

```python
import numpy as np
from thermoengine import phases
from thermoengine import model
import pyQ3
import dill
#%run core.ipynb
import importlib
import matplotlib.pyplot as plt
from dill import pickle

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
elements = ['O', 'H', 'Si', 'Na', 'Cl']
basis_species = ['H2O', 'H+', 'O2(G)', 'H4SIO4(AQ)', 'NAOH(AQ)', 'HCL(AQ)']
other_species = ['H6SI2O7(AQ)', 'H8SI3O10(AQ)', 'H3SIO4-', 'OH-', 'O2(AQ)', 'NACL(AQ)', 'NA+', 'CL-']
system_params = {'elements': elements,
                 'basis_species': basis_species,
                 'other_species': other_species,
                 'hydrated_species': pyQ3.defaultsystem.hydrated_species}
```

```python
with open('eq_working/dew_system.pkl','wb') as file:
    dill.dump(system_params, file)
    file.close()
```
